```python
import pandas as pd
import yfinance as yf
```

```python
symbols_table = pd.read_html("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies", header=0)[0]
```

```python
symbols_table
```

```python
symbols_table_new = symbols_table.set_index("Symbol")
```

```python
symbols_table_new
```

```python
symbols_table_new["Ticker Yahoo"] = symbols_table_new.index.str.replace(".", "-", regex=False)
```

```python
symbols_table_new
```

```python
symbols = list(symbols_table_new.loc[:, "Ticker Yahoo"])
```

```python
symbols
```

```python
new_data = yf.download(tickers = symbols,
                        start="2017-01-01",
                        group_by = "column",
                        interval = "1d",
                        auto_ajust = True,
                        threads = True,
                           )
```
